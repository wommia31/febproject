MACHINE LEARNING

LINEAR REGRESSION 

linear regression model

To work well with linear regression I had to drop columns that doesnt have numeric values

```python
malaria3.head(2)
```

```python
malaria4 = malaria3.drop(['Regions', 'Residence','Having health insurance','Economic status','state'], axis=1)
```

```python
malaria4.head(2)
```

Assigning X and Y array

```python
X = malaria4[["Unique identifeyer Code","Water source","Floor type","Wall type","Roof type","IPI in pregnancy","People using basic sanitation","children receiving AMD","Use of treated mosquito net","Having Bicycle"]]
y = malaria4["awareness on malaria"]
```

Training,testing and split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

Checking to ensure we have the the desired 70% train, 30% test split of the data.

```python
print("{0:0.2f}% in training set".format((len(X_train)/len(malaria4.index)) * 100))
print("{0:0.2f}% in test set".format((len(X_test)/len(malaria4.index)) * 100))
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

   -1 unit increase of water source is associated with a decrease in malaria awareness by -8.5.
   -1 unit increase in IPI in pregnancy is associated with decrease in malaria awareness by -2.6
   -1 unit increase of people using basic sanitation is associated with an increase in malaria awareness by 9.6
   -1 unit increase in children receiving AMD  is associated with an increase in malaria awareness
   -1 unit increase in use of treated mosquito nets  is associated with decrease in malaria awareness 
   -1 unit of increase in having bicycle is associated with increase in malaria awareness

Prediction

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=40,color="green");
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

Random Forest Model.
Its a technique that ensembles decison trees to create a more accurate model.used in classification and regression
Below-importing libraries

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42, n_estimators=10)      
rf_model.fit(X_train, y_train.ravel()) 
```

splitting data into trainig and testing

```python
rf_predict_train = rf_model.predict(X_train)
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_train, rf_predict_train)))
```

```python
rf_predict_test = rf_model.predict(X_test)
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, rf_predict_test)))
```

make prediction on testing data

```python
print(metrics.confusion_matrix(y_test, rf_predict_test) )
print("")
print("Classification Report")
print(metrics.classification_report(y_test, rf_predict_test))
```
